Let's import all these required modules: 

```python
# GraphLab 
import graphlab as gl
gl.canvas.set_target('ipynb') # output a single GraphLab Canvas view to a IPython Notebook cell

# Matplotlib
%matplotlib inline
import matplotlib.pyplot as mpl
mpl.rcParams['figure.figsize'] = (15.0, 8.0)

# Numpy
import numpy as np

# HSV values in [0..1)
# returns [r, g, b] values from 0 to 1, as required by matplotlib
def hsv_to_rgb(h, s=0.5,v=0.95):
    golden_ratio_conjugate = 0.618033988749895
    h += golden_ratio_conjugate
    h %= 1
    h_i = int((h*6))
    f = h*6 - h_i
    p = v * (1 - s)
    q = v * (1 - f*s)
    t = v * (1 - (1 - f) * s)
    if h_i == 0: r, g, b = v, t, p
    if h_i == 1: r, g, b = q, v, p 
    if h_i == 2: r, g, b = p, v, t 
    if h_i == 3: r, g, b = p, q, v 
    if h_i == 4: r, g, b = t, p, v 
    if h_i == 5: r, g, b = v, p, q
    return [r, g, b]
```

The **hsv_to_rgb** function is an utility function that will allow us to plot graphs with random but aesthetically pleasing colors. It takes three parameters that represent a color in the  [HSV](http://en.wikipedia.org/wiki/HSL_and_HSV) (Hue, Saturation, Value) space and it returns the corresponding [RGB](http://en.wikipedia.org/wiki/RGB) coding. 

This is convenient because we get equally bright and colorful colors by using a fixed value for saturation and value, and just modifying the hue. In this way, just generating a random number, the hue value, we obtain a random and colorful color.

## Data Exploration

It's time to explore the data in order to decide which information are useful to build our predictive models.

Let's start to analyze missing values for each column, plotting a stacked bar plot.

```python
columns = ("Pclass", "Name", "Sex", "Age", "SibSp", "Parch", "Ticket", "Fare", "Cabin", "Embarked")

not_null = [sum(1 for el in titanic_data[column] if el or el == 0) for column in columns]  
null = [len(titanic_data) - el for el in not_null] 

indexes = np.arange(len(columns)) # the x locations for the columns
width = 0.5 

not_null_bar = mpl.bar(indexes, not_null, width, color='green', edgecolor='white', alpha=0.8)
null_bar = mpl.bar(indexes, null, width, color='red', edgecolor='white', bottom=not_null, alpha=0.8)
mpl.xlim( indexes[0] - 0.5, indexes[-1] + 1)

mpl.title('Null values for each column', fontsize=20, weight='bold')
mpl.xticks(indexes + width/2., columns, fontsize=16)
mpl.yticks(np.arange(0,1200,100))
mpl.legend( (not_null_bar[0], null_bar[0]), ('Not Null', 'Null') )

# show percentage of missing values for each column
for ind in indexes:
    ind = int(ind)
    x = ind + width / 2.
    y = not_null[ind] + null[ind] + 10 
    percentage = null[ind] / float( not_null[ind] + null[ind]) * 100
    mpl.text(x, y, "%5.2f%%" % percentage, fontsize=16, ha='center')

mpl.show()
```

As we can see from the picture above, the **Cabin** column contains the majority of null values (*77.10%*), so we will ignore it for our predictive models. 

The **Age** column contains almost *20%* of missing values. This proportion is likely small enough for reasonable replacement with some form of imputation. Before we start filling in missing data, let's see what can be learned from the data we have. Putting some simple data visualization tools to work can take us a long way toward understanding what might influence the outcome we're trying to predict, i.e. whether or not a passenger survived.


### Passenger Class Distribution

Let's first print the distribution of the passenger class variable. Then, we further investigate how the proportion of survival changes for each class. 

We use the [show](https://graphlab.com/products/create/docs/generated/graphlab.SArray.show.html) method offered by the [SArray](https://graphlab.com/products/create/docs/generated/graphlab.SArray.html) structure, i.e. the **Pclass** column, in order to see the number of passengers for each class and plot a simple histogram.

```python
passenger_class = titanic_data["Pclass"].astype(str)
passenger_class.show()
```

The plot highlights that the majority of passengers were in the third class, the lower.

Let's investigate the proportion of survived for each class. In this case we use the [groupby](https://graphlab.com/products/create/docs/graphlab.data_structures.html#module-graphlab.aggregate) aggregator to aggregate passengers by their class (**Pclass**) and their destiny (**Survived**) and to count the number of passengers for each group. Finally, we use the [filter_by](https://graphlab.com/products/create/docs/generated/graphlab.SFrame.filter_by.html#graphlab.SFrame.filter_by) method in order to get the number of survived and the number of died in each class and plot the stacked bar plot. The percentage over each bar represents the percentage of survived of the corresponding passenger class.


```python
class_distribution = titanic_data.groupby(["Pclass", "Survived"], {'count':gl.aggregate.COUNT()})

survived = class_distribution.filter_by(1,'Survived').sort("Pclass")
died = class_distribution.filter_by(0,'Survived').sort("Pclass")

width = 0.5 

survived_bar = mpl.bar(survived["Pclass"], survived["count"], width, color='green', edgecolor='white', alpha=0.8)
died_bar = mpl.bar(died["Pclass"], died["count"], width, color='red', edgecolor='white', bottom=survived["count"], alpha=0.8)
mpl.xlim( indexes[0] - 0.5, indexes[-1] + 1)

mpl.title('Survived by Passengers Class', fontsize=20, weight='bold')
mpl.xticks(survived["Pclass"] + width/2., survived["Pclass"], fontsize=16)
mpl.xlim(0.5,4)
mpl.yticks(np.arange(0,600,50))
mpl.legend( (survived_bar[0], died_bar[0]), ('Survived', 'Died') )

for ind in np.arange(len(survived)):
    ind = int(ind)
    x = 1 + ind + width / 2.
    y = survived["count"][ind] + died["count"][ind] + 10 
    percentage = survived["count"][ind] / float( survived["count"][ind] + died["count"][ind]) * 100
    mpl.text(x, y, "%5.2f%%" % percentage, fontsize=20, ha='center')

mpl.show()
```

Looking the graph above, we can easily see that rich people were more likely to survive with respect to the poorest ones. Indeed, the *62.96%* of passengers in the first class have survived, whereas this percentage is just *24.24%* for people in the third class. 

This fact points out the the passenger class can be a good variable to use in our predictive models because is quite discriminative if we know the class of a given passenger.

### Age Distribution

Since the *Age* column contains *20%* of missing values, we use the [dropna](https://graphlab.com/products/create/docs/generated/graphlab.SFrame.dropna.html#graphlab.SFrame.dropna) function over the **Age** field of our **titanic_data** SFrame in order to remove all of them. Indeed, *dropna* allows to remove *'NA'* values (*None* or *NaN*) from one column and even beyond.

```python
age = titanic_data["Age"].dropna()
age.show()
```

The majority of passengers were between 17 and 40 years old. 
In what follows we use boxplot to analyze differences between ages among passengers who survived and who not.

```python
survived = titanic_data.filter_by(1,'Survived')["Age"].dropna()
died = titanic_data.filter_by(0,'Survived')["Age"].dropna()

data_to_plot = [died, survived]

bp = mpl.boxplot(data_to_plot,patch_artist=True)

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#1b9e77' )

## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
    

mpl.xticks([1,2],['Died', 'Survived'], fontsize=20)
mpl.yticks(np.arange(0,85,5))
mpl.title("Survived by Age", fontsize=20, weight='bold')
mpl.show()
```

From the boxplot above there is no much difference between the age of who died and who survived. 
So the theory of ["Survival of the fittest"](http://en.wikipedia.org/wiki/Survival_of_the_fittest) does not hold here. However, let's further analyze if the *"Women and children first"* protocol is respected also for child and not only for women.

In order to do this we create and add a new column to our SFrame, called **Categorized_age**. 
We do this starting from the **Age** column and invoking the [apply](https://graphlab.com/products/create/docs/generated/graphlab.SArray.apply.html) method to apply a function to each element. We categorize all passengers below 10 years as *Child* and as *Adult* those who remain.

```python
titanic_data['Categorized_Age'] = titanic_data['Age'].apply(lambda x: "Child" if x <= 10 else "Adult")

age_distribution = titanic_data.groupby(["Categorized_Age", "Survived"], {'count':gl.aggregate.COUNT()}).dropna()

survived = age_distribution.filter_by(1,'Survived').sort("Categorized_Age")
died = age_distribution.filter_by(0,'Survived').sort("Categorized_Age")

indexes = np.arange(len(survived["Categorized_Age"]))


width = 0.5 

survived_bar = mpl.bar(indexes, survived["count"], width, color='green', edgecolor='white', alpha=0.8)
died_bar = mpl.bar(indexes, died["count"], width, color='red', edgecolor='white', bottom=survived["count"], alpha=0.8)
mpl.xlim( indexes[0] - 0.5, indexes[-1] + 1)

mpl.title('Survived by Age Categorization', fontsize=20, weight='bold')
survived["Categorized_Age"] = [sex.capitalize() for sex in survived["Categorized_Age"]]
mpl.xticks(indexes + width/2., survived["Categorized_Age"], fontsize=16)
mpl.xlim(-0.5,2)
mpl.yticks(np.arange(0,700, 50))
mpl.legend( (survived_bar[0], died_bar[0]), ('Survived', 'Died') )

for ind in indexes:
    ind = int(ind)
    x = ind + width / 2.
    y = survived["count"][ind] + died["count"][ind] + 10 
    percentage = survived["count"][ind] / float( survived["count"][ind] + died["count"][ind]) * 100
    mpl.text(x, y, "%5.2f%%" % percentage, fontsize=20, ha='center')


mpl.show()
```

The *59.38%* of children have survived with respet to just *38.77%* of adults, so the age can still have a role for our prediction. 

### Siblings/Spouses Aboard Distribution

It's time to see the distribution of the number of siblings and/or spouses of each passenger and see how the percentage to survive
change according to this number.

```python
sibling_spouses = titanic_data["SibSp"].astype(str)
sibling_spouses.show()
```

The majority of Titanic passengers have no sibling/spouse and 209 of them have one sibling or spouse.

In the code below we genarate the stacked bar plot analyzing the proportion of survived passengers according to the number of sibling/spouses. The first loop **for** (line 6) is just for plotting purpose. Indeed, the **bar** method (line 12,13) expects that we have the number of survived/died passenger for each number of sibling/spouses but, however, in the dataset there are no passengers that have survived with 5 or 8 siblings/spouses. For this reason, we append in the SFrame (line 8) a line for these cases with the *count* variable equals to zero.

```python
sibsp_distribution = titanic_data.groupby(["SibSp", "Survived"], {'count':gl.aggregate.COUNT()}).sort(["SibSp"])

survived = sibsp_distribution.filter_by(1,"Survived")
died = sibsp_distribution.filter_by(0,"Survived")

for sibsp in sibsp_distribution["SibSp"]:
    if not survived.filter_by(sibsp, "SibSp"):
        survived = survived.append(gl.SFrame({'SibSp': [sibsp], 'Survived': [1], 'count':[0]}))
        
width = 0.5 

survived_bar = mpl.bar(survived["SibSp"], survived["count"], width, color='green', edgecolor='white', alpha=0.8)
died_bar = mpl.bar(died["SibSp"], died["count"], width, color='red', edgecolor='white', bottom=survived["count"], alpha=0.8)
mpl.xlim( indexes[0] - 0.5, indexes[-1] + 1)

mpl.title('Survived by SibSp', fontsize=20, weight='bold')
mpl.xticks(survived["SibSp"] + width/2., survived["SibSp"], fontsize=16)
mpl.xlim(-0.5,9)
mpl.yticks(np.arange(0,750,50))
mpl.xlabel("SibSp", fontsize=16)
mpl.legend( (survived_bar[0], died_bar[0]), ('Survived', 'Died') )

for ind in np.arange(len(survived)):
    ind = int(ind)
    x = survived["SibSp"][ind] + width / 2.
    y = survived["count"][ind] + died["count"][ind] + 10 
    percentage = survived["count"][ind] / float( survived["count"][ind] + died["count"][ind]) * 100
    mpl.text(x, y, "%5.2f%%" % percentage, fontsize=20, ha='center')

mpl.show()
```

People with one sibling/spouse have more probabilities to survive with respect to those with neither a sibling nor a spouse.
However, there are few people with more than 2 siblings/spouses so this variable have less influence with respect to the previous ones.

# Model Building

After doing some exploratory analysis of the data, we now need to clean the dataset to create our models. Note that it is important to explore the data so that we understand what elements need to be cleaned. For example you might have noticed that there are missing values in the data set, especially in the **Age** column.

A common approach to tackle missing values is to replace them with the average of the available values. Using the [sketch_summary](https://graphlab.com/products/create/docs/generated/graphlab.SArray.sketch_summary.html#graphlab.SArray.sketch_summary) metod over the **Age** column, we can query many descriptive statistics on this **SArray**.

```python
titanic_data["Age"].sketch_summary()
```

From the previous command we could replace all the missing values for age with the mean, i.e. *29.7*. However, taking this approach would be fine if only a small fraction of the ages were missing. With missings accounting for 20 percent of all **Age** data in a relatively small data set (less then 900 records), we could look for a more refined method of imputation.

Indeed, we can use a smarter way to replace missing values. One intuition can be that the age can vary considering the type of people, e.g. the discovery of ["The Richer You Are the Older You’ll Get"](http://blogs.wsj.com/economics/2014/04/18/the-richer-you-are-the-older-youll-get/). We can capture this considering the passenger class. Let's analyse how age changes according the passenger class with a simple boxplot:


```python
classes = titanic_data["Pclass"].unique().sort()

data_to_plot = [ titanic_data.filter_by(pclass,'Pclass')["Age"].dropna() for pclass in classes]

bp = mpl.boxplot(data_to_plot,patch_artist=True)

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#1b9e77' )

## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
    

mpl.title("Age by Passenger Class", fontsize=20, weight='bold')
mpl.yticks(np.arange(-5,100,5))
mpl.show()
```

As we can see, the median age of the first class is higher than that of the second one which, in turn, is higher than that of the third one. So it is better to consider this fact, replacing the missing values according to the passenger class.

We can also improve this imputation strategy by looking closer the **Name** field, using the [head](https://graphlab.com/products/create/docs/generated/graphlab.SFrame.head.html#graphlab.SFrame.head) method:

```python
titanic_data[["PassengerId","Name"]].head()
```

Notice the titles, *Mr., Mrs., Miss., Master*, following each of the surnames. The passenger's title should help with differentiation between younger and older females/male. Indeed, a *Miss.* is usually younger than a *Mrs.* and a *Master* is usually younger than a *Mr.*. According to the [English honorifics](http://en.wikipedia.org/wiki/English_honorific), the title *Master* is used for young men and boys. Finally, we can also assume that people with the same titles are closer together in age.

So in order to impute the age for missing values, we will use both the passengers class and their title. In order to do that, we will create a new column, called **Title**.

```python
def get_title(name):
    start = name.find(", ") + len(", ")
    end = name.find(". ") + 1
    return name[start:end] 

titanic_data["Title"] = titanic_data["Name"].apply(lambda x: get_title(x))

titanic_data.head()
```

Let's see all the unique titles present in the dataset using the **unique** method:

```python
titanic_data["Title"].unique()
```

It's time now to impute the missing values for the **Age** column.

```python
age_for_imputation = titanic_data.groupby(["Pclass","Title"],{'Median':gl.aggregate.QUANTILE('Age', 0.5)})

def impute_age(pclass, title):
    return age_for_imputation["Median"][(age_for_imputation["Pclass"] == pclass) & 
                                         (age_for_imputation["Title"] == title)][0][0]

titanic_data["Age"] = [row["Age"] if row["Age"] != None else impute_age(row["Pclass"], row["Title"]) 
                                                                for row in titanic_data[["Age", "Pclass", "Title"]]]
```

We may recall that the **Embarked** feature also had *0.22%* of missing values. 
We can see a summary of this feature using the **show** method.

```python
port_of_embarkation = titanic_data["Embarked"]
port_of_embarkation.show()
```

The graph above reveals just two missings. It should be fine to replace those missings with the most common value, i.e. *S (Southampton)*.

```python
titanic_data["Embarked"] = titanic_data["Embarked"].apply(lambda x: x if x != '' else "S")
port_of_embarkation = titanic_data["Embarked"]
port_of_embarkation.tail(1)
port_of_embarkation.show()
```

#### Classifiers

Classifiers are very easy to build with GraphLab. We have just to use the [create](https://graphlab.com/products/create/docs/graphlab.toolkits.classifier.html) method that requires the training set and the set of features we want to use. Let's start to build a very simple model, considering just the *Sex* feature.

```python
model_1 = gl.classifier.create(train_set, target='Survived', features=['Sex'])
print model_1.summary()
```

We can evaluate the goodness of the model using the *[evaluate](https://graphlab.com/products/create/docs/generated/graphlab.logistic_regression.LogisticRegressionModel.evaluate.html#graphlab.logistic_regression.LogisticRegressionModel.evaluate)* method, that requires the test set as parameter.
This metod returns the [accuracy](http://en.wikipedia.org/wiki/Accuracy_and_precision) and the [confusion matrix](http://en.wikipedia.org/wiki/Confusion_matrix). With the last one we can compute other useful metrics, like [precision](http://en.wikipedia.org/wiki/Precision_and_recall), [recall](http://en.wikipedia.org/wiki/Precision_and_recall), and the [F-1 score](http://en.wikipedia.org/wiki/F1_score). 

```python
result_1 = model_1.evaluate(test_set)

def get_precision(confusion_matrix):
    return confusion_matrix['true_positive'] / float(confusion_matrix['true_positive'] + confusion_matrix['false_positive'])

def get_recall(confusion_matrix):
    return confusion_matrix['true_positive'] / float(confusion_matrix['true_positive'] + confusion_matrix['false_negative'])

def get_f1_score(confusion_matrix):
    precision = get_precision(confusion_matrix) 
    recall = get_recall(confusion_matrix)
    return 2.0 * (precision * recall) / (precision + recall)

def print_statistics(result):
    print "*" * 30
    print "Accuracy        : ", result["accuracy"]
    print "Confusion Matrix: \n", result["confusion_matrix"]    

print_statistics(result_1)
```

Just using the **Sex** feature we obtain reasonable accuracy. Note that different executions of this notebook can achieve slightly different results of the models since we use a random selection for train/test datasets and the dataset is small. Let's try to create a new model considering other features, like the passenger class and their age that seem quite discriminative. 

```python
model_2 = gl.classifier.create(train_set, target='Survived', features=['Sex', 'Age', 'Pclass'])

result_2 = model_2.evaluate(test_set)
print_statistics(result_2)
```
